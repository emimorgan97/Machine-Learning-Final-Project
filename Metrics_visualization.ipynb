```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Metrics for each model
models = ['TF Model', 'Attention Model', 'Dense Net Model']
true_positives = [5580, 9092, 3641]
true_negatives = [11309, 10518, 4416]
false_positives = [5082, 5873, 573]
false_negatives = [10797, 7285, 1370]
```

```python
def plot_true_positives_and_negatives():
    x = np.arange(len(models))
    width = 0.35

    fig, ax = plt.subplots()
    bars1 = ax.bar(x - width/2, true_positives, width, label='True Positives')
    bars2 = ax.bar(x + width/2, true_negatives, width, label='True Negatives')

    ax.set_ylabel('Counts')
    ax.set_title('True Positives and True Negatives by Model')
    ax.set_xticks(x)
    ax.set_xticklabels(models)
    ax.legend()

    plt.show()
```

```python
def calculate_metrics(tp, tn, fp, fn):
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    return accuracy, precision, recall, f1_score
```

```python
def display_metrics():
    for i in range(len(models)):
        accuracy, precision, recall, f1_score = calculate_metrics(true_positives[i], true_negatives[i], false_positives[i], false_negatives[i])
        print(f"\nMetrics for {models[i]}:")
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1 Score: {f1_score:.4f}")
```

```python
def plot_metrics_comparison():
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    all_metrics = []

    for i in range(len(models)):
        accuracy, precision, recall, f1_score = calculate_metrics(true_positives[i], true_negatives[i], false_positives[i], false_negatives[i])
        all_metrics.append([accuracy, precision, recall, f1_score])

    all_metrics = np.array(all_metrics)

    fig, ax = plt.subplots()

    bar_width = 0.2
    opacity = 0.4
    colors = ['b', 'g', 'r', 'y']

    for i in range(len(metrics)):
        ax.bar(np.arange(len(models)) + i * bar_width, all_metrics[:, i], bar_width,
               alpha=opacity, color=colors[i], label=metrics[i])

    ax.set_xlabel('Models')
    ax.set_ylabel('Scores')
    ax.set_title('Model Performance Metrics Comparison')
    ax.set_xticks(np.arange(len(models)) + bar_width)
    ax.set_xticklabels(models)
    ax.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Display metrics
display_metrics()

# Plot true positives and true negatives
plot_true_positives_and_negatives()

# Plot metrics comparison
plot_metrics_comparison()
```
